##Regarding the homework assignment

Try to limit the amount of code repetition. It's not a competition for the least number of lines of code, but don't duplicate unnecessarily.

```
def letter_grade(score, in_pct = True, pf = False):
    '''returns letter grade based on provided score'''
    # if in_pct is false, we just multiply the score by 100, so 0.90 -> 90%
    if in_pct == False:
        score *= 100
    # we use two parallel lists of letter grades and associated lower thresholds
    letter_grades = ['A', 'B', 'C', 'D']
    thresholds = [90, 80, 70, 60]
    # if we don't get a letter grade later on, we will just return a failing grade
    lg='F'
    # loop over both letter grades and thresholds simultaneously
    for grade,thresh in zip(letter_grades,thresholds):
        # if we have a score above this threshold, then we get this grade
        if score>=thresh:
            lg = grade
            # no need to continue, so we break out of the for loop
            break
    # if pass/fail, just return pass instead of the letter grade
    if pf == True  and lg!='F':
        return 'P'
    return lg

# loop over all scores from 100 to 50, downwards
for score in range(100,50,-1):
    print(score,
          letter_grade(score),
          letter_grade(score, pf = True),
          letter_grade(score/100, in_pct = False))
```

###Example: recursive factorial

```
def my_factorial_recursive(n):
    ''' compute n! recursively '''
    if n == 0:
        return 1 # since 0! = 1
    else:
        return n * my_factorial_recursive(n - 1)

for n in range(0, 10):
    result = my_factorial_recursive(n)
    print("n =", n, "  n! =", result)
```

```
def A(m,n):
  '''return Ackermann function for m,n'''
  if m < 0 or n < 0:
    print ("ERROR: you must call this function with m > 0, n > 0")
    return 0 
  if m == 0:
    return n + 1
  elif m > 0 and n == 0:
    return A(m-1, 1)
  else:
    return A(m-1, A(m,n-1))
  
print(A(1.5, 1.5))
```

#### Possible solution

```
def A(m,n):
    ''' compute A(m,n) recursively '''
    #print("  A(",m,",",n,")")
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return A(m-1,1)
    elif m > 0 and n > 0:
        return A(m-1,A(m,n-1))

for n in range(0,3):
  for m in range(0,4):
    print("A(",m,",",n,") = ",A(m,n))
```

##Exercise: a sorting algorithm

**Goal**: write a function to sort a list of integers into ascending order (1, 2, 3....).

**To get started**:  Sort the following by hand. Pay attention to the steps you are taking.

```
L = [55, 11,  8, 56, 39, 42, 17,  9, 86, 26]
```

###Bubble sort

**Key ideas:**
* Loop through list, comparing and swapping (if needed) adjacent entries
* Loop repeatedly until sorted.
* Sorted = a loop through the list in which no swaps. 

[YouTube video](https://www.youtube.com/watch?v=Cq7SMsQBEUw)

```
L = [55, 11,  8, 56, 39, 42, 17,  9, 86, 26]

def mysort2(L):
    ''' bubble sort real numbers in ascending order'''
    is_sorted = False
    while not is_sorted:
        nswap = 0
        for i in range(1, len(L)):
            if L[i] < L[i-1]:
                nswap = nswap + 1
                L[i], L[i-1] = L[i-1], L[i]
        if nswap == 0: 
            is_sorted = True
    return L
            
mysort2(L)
```

###Insertion Sort

**Key ideas:**
* Gradually sort list from "left" to "right"
* When sorting in element `i_to_sort` take advantage of the fact that `[0,...,i_to_sort-1]` is already sorted

[YouTube video](https://www.youtube.com/watch?v=8oJS1BMKE64)

```
L = [55, 11,  8, 56, 39, 42, 17,  9, 86, 26]

def mysort(L):
    ''' Insertion sort real numbers in ascending order '''
    for i_to_sort in range(1, len(L)): # sort in a new element with index i_to_sort
        for icompare in range(i_to_sort, 0, -1):
            if L[icompare-1] > L[icompare]:
                L[icompare-1], L[icompare] = L[icompare], L[icompare-1]
            else:
                # this takes advantage of 0 ... i_to_sort-1 
                # already being sorted
                # now, get out of inner (icompare) loop
                break  
        # end of i_to_sort loop
        # i_to_sort will be incremented for the next iteration
    return L

mysort(L)
```

###Merge sort (recursive)

** Key ideas: **
* Divide and conquer
* Divide list roughly in half to make `A` and `B`
  * Then use _recursion_ to divide again, and again, and again to get `A'`,`B'`, `A''`, `B''`, etc
* Eventually, we get to 1 or 2 element lists
  * Start merging to build up sorted lists
  * Merging takes advantage of `A'''` and `B'''` being sorted
* Clever and tricky, helps to watch it work
* Invented by physics/math/comp-sci demi-god _John von Neumann_

[YouTube video](https://www.youtube.com/watch?v=ZRPoEKHXTJg)

```
L = [55, 11,  8, 56, 39, 42, 17,  9, 86, 26]

def mysort3(L):
    ''' merge sort numbers in ascending order'''
    if len(L)<2 : return L
    q = len(L)//2 # intentional integer division
    A = mysort3(L[:q])
    B = mysort3(L[q:])
    return mymerge(A,B)

def mymerge(A,B):
    ''' merge two already sorted lists '''
    i, j = 0, 0
    C = []
    while True:
        if A[i] < B[j]:
            C.append(A[i])
            i += 1
        else: # B[j] <= A
            C.append(B[j])
            j += 1
        if i == len(A): # this is where we take advantage of A, B being already sorted
            C += B[j:]
            return C
        elif j == len(B): # here too
            C += A[i:]
            return C

mysort3(L)
```
